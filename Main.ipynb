```python
import pandas as pd
import numpy as np

ratings = pd.read_csv('tmdb_5000_credits.csv')
ratings.columns
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')

ratings = pd.merge(movies,ratings)
ratings.columns
```

```python
ratings.drop({'movie_id','budget','genres','homepage','keywords',
              'original_language','original_title','overview','popularity',
              'production_companies','production_countries','release_date','revenue','runtime',
              'spoken_languages','status','tagline','vote_count','cast','crew'},axis=1,inplace=True)
ratings
```

```python
user_ratings = ratings.pivot_table(index=['id'],columns=['title'],values='vote_average')
user_ratings
```

```python
user_ratings = user_ratings.fillna(0)
user_ratings
```

```python
item_similarity_df = user_ratings.corr(method='pearson')
item_similarity_df.head(100)
```

```python
def get_similar_movies(movie_name,user_rating):
    similar_score = item_similarity_df[movie_name]*(user_rating-5)
    similar_score = similar_score.sort_values(ascending=False)

    return similar_score
```

```python
#TestData
import random as rand


li = list()
for j in range(1,100000):
    ni=rand.randint(1,10)
    rand_user = set()
    for i in range(ni):
        user1 = ratings.sample(axis='index')
        user1 = user1.title
        rate = rand.randint(0,10)
        user1 = user1.values
        rand_user.add((user1[0],rate))
    li.append(rand_user)

li.__len__()
```

```python
import pickle
pickle.dump(li,open('listofusers.pkl','wb'))
```

```python
similar_movies = pd.DataFrame()

val = int(input("Enter the userid(0-100000):"))
rand_user = li[val]
for movie,rating in rand_user:
    similar_movies = similar_movies.append(get_similar_movies(movie,rating),ignore_index=True)

similar_movies.head()
z=similar_movies.sum().sort_values(ascending=False)

z=z.to_frame()
z=z.drop(columns=0)
pickle.dump(z.head(5),open('results.pkl','wb'))
z
```
